```python
# ==============================
# STEP 1: Install & Import Libraries
# ==============================
import pandas as pd
import sqlite3

# ==============================
# STEP 2: Upload Excel File
# ==============================
from google.colab import files
uploaded = files.upload()

# Replace with your actual file name after upload
file_name = list(uploaded.keys())[0]

# ==============================
# STEP 3: Load Dataset
# ==============================
df = pd.read_excel('/content/drive/MyDrive/Decodelabs/Dataset for Data Analytics2.xlsx')

print("Dataset Preview:")
print(df.head())

print("\nColumn Names:")
print(df.columns)

# ==============================
# STEP 4: Create SQLite Database
# ==============================
conn = sqlite3.connect(':memory:')
df.to_sql('data', conn, index=False, if_exists='replace')

# ==============================
# STEP 5: Run SQL Queries
# ==============================

# 1. View all data
print("\n1. All Data:")
print(pd.read_sql_query("SELECT * FROM data", conn))


# 2. Filter (example: change column name if needed)
print("\n2. Filter Example:")
print(pd.read_sql_query("SELECT * FROM data WHERE 1=1 LIMIT 5", conn))


# 3. Order By (Top values)
print("\n3. Ordered Data:")
print(pd.read_sql_query("SELECT * FROM data ORDER BY 1 DESC LIMIT 5", conn))


# 4. COUNT
print("\n4. Count:")
print(pd.read_sql_query("SELECT COUNT(*) AS total_rows FROM data", conn))


# 5. SUM (replace 'amount' with numeric column)
numeric_col = df.select_dtypes(include='number').columns[0]
print("\n5. SUM:")
query = f"SELECT SUM({numeric_col}) AS total_sum FROM data"
print(pd.read_sql_query(query, conn))


# 6. AVG
print("\n6. AVG:")
query = f"SELECT AVG({numeric_col}) AS avg_value FROM data"
print(pd.read_sql_query(query, conn))


# 7. GROUP BY (use first categorical column)
cat_col = df.select_dtypes(include='object').columns[0]

print("\n7. GROUP BY:")
query = f"""
SELECT {cat_col}, COUNT(*) as count_records
FROM data
GROUP BY {cat_col}
"""
print(pd.read_sql_query(query, conn))


# 8. GROUP BY + SUM
print("\n8. GROUP BY + SUM:")
query = f"""
SELECT {cat_col}, SUM({numeric_col}) as total_value
FROM data
GROUP BY {cat_col}
"""
print(pd.read_sql_query(query, conn))


# 9. HAVING (advanced)
print("\n9. HAVING:")
query = f"""
SELECT {cat_col}, SUM({numeric_col}) as total_value
FROM data
GROUP BY {cat_col}
HAVING total_value > 1000
"""
print(pd.read_sql_query(query, conn))


# 10. ORDER BY with GROUP BY
print("\n10. GROUP BY + ORDER BY:")
query = f"""
SELECT {cat_col}, SUM({numeric_col}) as total_value
FROM data
GROUP BY {cat_col}
ORDER BY total_value DESC
"""
print(pd.read_sql_query(query, conn))


# ==============================
# STEP 6: Close Connection
# ==============================
conn.close()
```
